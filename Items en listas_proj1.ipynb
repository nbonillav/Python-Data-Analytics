Crear un menu din√°mico, que pregunte hasta indicar lo contrario si quiere agregar un item a las lista:
Lista de series
Lista de Canciones
Lista de Juegos

```python
def menu():
    print("-------------------------")
    print("Hola!")
    name = input("Cual es su nombre? ")
    print("-------------------------")

    def open_append_close(categ):
        category = "my-" + str(categ) + ".txt"
        # print(category)
        file = open(category, "a")
        if categ == 'songs':
            insert_stuff = input("Agregue una Canci√≥n: ")
        elif categ == 'series':
            insert_stuff = input("Agregue una Serie: ")
        else:
            insert_stuff = input("Agregue un Juego: ")
        print('---------------------------------\n')
        file.write(insert_stuff + "\n")
        file.close()
        return insert_stuff

    def read_files():
        categories = ['songs', 'series', 'games']
        for i in categories:
            category = "my-" + str(i) + ".txt"
            file = open(category, "r+")
            print('---------------------------------')
            print(category)
            print('---------------------------------')
            print(file.read())


    while(True):
        print("Bienvenid@", name)
        print("1. Agregar a lista de Canciones")
        print("2. Agregar a lista de Series")
        print("3. Agregar a lista de Juegos")
        print("4. Salir")
        answer = input("Elija una opcion: ")

        if answer == "1":
            open_append_close('songs')
        elif answer == "2":
            open_append_close('series')
        elif answer == "3":
            open_append_close('games')
        elif answer == "4":
            break
        else:
            read_files()

menu()
```
