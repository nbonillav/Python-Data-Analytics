## Practica 2

1. Seleccione las variables "diversity", "in_state_tuition", "out_of_state_total" y "region". Genere una nueva variable con estas columnas.
2. Ordene el dataset nuevo de manera descendente, a partir de la variable "in_state_tuition".
3. Filtre el dataset cuando la variable "region" sea "South" y "West".
4. Filtre el dataset cuando la variable "out_of_state_total" sea menor a 9000 y "diversity" sea igual a "high".
5. Genere una nueva columna que tenga la división entre las variables "in_state_tuition" y "out_of_state_total"

```python
# 1.Seleccione las variables "diversity", "in_state_tuition", "out_of_state_total" y "region". Genere una nueva variable con estas columnas.
new_var = df[["diversity", "in_state_tuition", "out_of_state_total", "region"]]
new_var
```

```python
#2.Ordene el dataset nuevo de manera descendente, a partir de la variable "in_state_tuition".
new_var.sort_values("in_state_tuition", ascending = False)
```

```python
#3.Filtre el dataset cuando la variable "region" sea "South" y "West".
new_var[new_var["region"].isin(["South", "West"])]
```

```python
# 4.Filtre el dataset cuando la variable "out_of_state_total" sea menor a 9000 y "diversity" sea igual a "high".
new_var[ (new_var["out_of_state_total"] > 9000) & (new_var["diversity"] == "high")]
```

```python
# 5.Genere una nueva columna que tenga la división entre las variables "in_state_tuition" y "out_of_state_total"
new_var["division"] = new_var["in_state_tuition"] / new_var["out_of_state_total"]
new_var
```
